```python
import pandas as pd
import sqlite3 as sql
```

```python
database = "chinook.db"
conn = sql.connect(database)
```

```python
q = """ select * from sqlite_master where type = "table"  """
pd.read_sql_query(q,conn)
```

```python
def query(q):
    return pd.read_sql_query(q,conn)
```

### customer_name, no_of_visit, total_spend

```python
q = """ select c.first_name || " " || c.last_name "Customer Name", i.invoice_id "No of Visits", sum(i.total) "Total Spend"
        from customer c
        inner join invoice i
        on
        c.customer_id=i.customer_id
        group by 1
        order by 3 DESC"""
query(q)
```

```python
q = """ select c.first_name || " " || c.last_name "Customer Name", i.invoice_id "No of Visits", sum(i.total) "Total Spend",
        case
            when sum(i.total) > 100 then "Big Customer"
            when sum(i.total) < 40 then "Small Customer"
            else "Regular Customer"
        end "Customer Type"
        
        from customer c
        inner join invoice i
        on
        c.customer_id=i.customer_id
        group by 1
        """
query(q)
```

```python
q = """ with customer_record as
        (select c.first_name || " " || c.last_name "Customer Name", i.invoice_id "No of Visits", sum(i.total) "Total Spend",
        case
            when sum(i.total) > 100 then "Big Customer"
            when sum(i.total) < 40 then "Small Customer"
            else "Regular Customer"
        end "Customer Type"
        
        from customer c
        inner join invoice i
        on
        c.customer_id=i.customer_id
        group by 1)
        
        select "Customer Type" ,count("Customer Type") Quantity from customer_record
        group by 1
        """
query(q)
```

### Customer who are regular

```python
q = """ select c.first_name || " " || c.last_name "Customer Name", i.invoice_id "No of Visits", sum(i.total) "Total Spend"
        from customer c
        inner join invoice i
        on
        c.customer_id=i.customer_id
        group by 1
        having "Total Spend" < 100 and "Total Spend" > 40
        """
query(q)
```

## Creating View

```python
q = """ create view Regular_Customer as
        select c.first_name || " " || c.last_name "Customer Name", i.invoice_id "No of Visits", sum(i.total) "Total Spend"
        from customer c
        inner join invoice i
        on
        c.customer_id=i.customer_id
        group by 1
        having "Total Spend" < 100 and "Total Spend" > 40
        """
conn.execute(q)
```

## Calling view

```python
q = """ select * from sqlite_master where type = "view" """
query(q)
```

```python
q = """ select * from Regular_Customer"""
query(q)
```
